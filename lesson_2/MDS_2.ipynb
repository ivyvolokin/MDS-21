```python
from math import *
A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
 ]

A = np.array(A) 
x = A[:,0] 
y = A[:,1]
type(A)
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
# Повернуть на 130 градусов
B = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
 ]

B = np.array(B) 
x = B[:,0] * cos(130) - B[:,1] * sin(130)
y = B[:,0] * sin(130) + B[:,1] * cos(130)

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
# Отразить объект относительно прямой y=x
C = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
 ]

C = np.array(C) 
y = C[:,0]    #замена осей на противоположенные
x = C[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
```

```python
# Выполнить смещение объекта и масштабирование (применить 2 матрицы трансформаций). 
# Смещение на (200, 300) и масштабирование на (0.5, 1.2)
D = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
 ]
D = np.array(D) 
sm = [
[200, 300, 0],
[200, 300, 0],
[200, 300, 0],
[200, 300, 0],
[200, 300, 0],
[200, 300, 0]
]
sm = np.array(sm)  #Смещение на (200, 300)

D = D + sm

vm = np.array([0.5, 1.2, 1]) #масштабирование на (0.5, 1.2)

D = D * vm

x = D[:,0] 
y = D[:,1]

plt.plot(x, y)
plt.ylim([200, 500])
plt.xlim([0, 200])
```
